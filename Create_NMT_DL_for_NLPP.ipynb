## **16k model**

```
# apply and create vocab train
! subword-nmt apply-bpe -c ./codes/codes_files_16k.bpe --vocabulary ./transformer_model_16k/vocab_16k.fr --vocabulary-threshold 50 < ./drive/MyDrive/train.fr > ./transformer_model_16k/train.BPE.fr
! subword-nmt apply-bpe -c ./codes/codes_files_16k.bpe --vocabulary ./transformer_model_16k/vocab_16k.en --vocabulary-threshold 50 < ./drive/MyDrive/train.en > ./transformer_model_16k/train.BPE.en

# Now the dev 
! subword-nmt apply-bpe -c ./codes/codes_files_16k.bpe --vocabulary ./transformer_model_16k/vocab_16k.fr --vocabulary-threshold 50 < ./drive/MyDrive/dev.fr > ./transformer_model_16k/dev.BPE.fr
! subword-nmt apply-bpe -c ./codes/codes_files_16k.bpe --vocabulary ./transformer_model_16k/vocab_16k.en --vocabulary-threshold 50 < ./drive/MyDrive/dev.en > ./transformer_model_16k/dev.BPE.en

# Now the test
! subword-nmt apply-bpe -c ./codes/codes_files_16k.bpe --vocabulary ./transformer_model_16k/vocab_16k.fr --vocabulary-threshold 50 < ./drive/MyDrive/test.fr > ./transformer_model_16k/test.BPE.fr
! subword-nmt apply-bpe -c ./codes/codes_files_16k.bpe --vocabulary ./transformer_model_16k/vocab_16k.en --vocabulary-threshold 50 < ./drive/MyDrive/test.en > ./transformer_model_16k/test.BPE.en

# vocab saving
!onmt-build-vocab --size 50000 --save_vocab ./transformer_model_16k/vocab.fr ./transformer_model_16k/train.BPE.fr
!onmt-build-vocab --size 50000 --save_vocab ./transformer_model_16k/vocab.en ./transformer_model_16k/train.BPE.en
```
